```python
%%javascript

import Root from '../../../../../../tree/treezjs/src/root/root.js';
import ChordMode from '../../../../../../tree/treezjs/src/result/chord/chordMode.js';
import Color from '../../../../../../tree/treezjs/src/components/color/color.js';

window.createModel = function(){

    let root = new Root();

    //#region STUDIES -----------------------------------------------------
    let studies = root.createStudies();
    //#end region ---------------------------------------------------------

    //#region RESULTS -----------------------------------------------------
    let results = root.createResults();

    //#region data

    let data = results.createData();
    let table = data.createTable();
    let columns = table.createColumnFolder('columns');
    columns.createColumn('source');
    columns.createColumn('target');
    columns.createColumn('value');

    table.createRow(["DE", "FR", "1"]);
    table.createRow(["DE", "CH", "2"]);
    table.createRow(["DE", "NL", "3"]);
    table.createRow(["FR", "CH", "10"]);
    table.createRow(["NL", "CH", "20"]);

    //#end region

    //#region page

    let page = results.createPage();
    let graph = page.createGraph();
    let chord = graph.createChord();
    chord.data.sourceData = 'root.results.data.table.columns.source';
    chord.data.targetData = 'root.results.data.table.columns.target';
    chord.data.valueData = 'root.results.data.table.columns.value';
    chord.data.legendText = 'My legend text';

    chord.nodes.innerRadius = '3.5 cm';

    chord.nodeLabels.textDistance = '-14 px';
    chord.nodeLabels.isAzimuthal = true;
    chord.nodeLabels.imageDistance = '20 px';

    chord.links.fillMode = ChordMode.value;

    let dE = chord.createChordNode('DE');
    dE.description = 'Germany';
    dE.color = Color.grey;
    dE.svg = '<svg width="3%" height="3%" viewBox="0 0 640 480">\n  <path fill="#ffce00" d="M0 320h640v160H0z"/>\n  <path d="M0 0h640v160H0z"/>\n  <path fill="#d00" d="M0 160h640v160H0z"/>\n</svg>';

    let fR = chord.createChordNode('FR');
    fR.description = 'France';
    fR.color = Color.blue;
    fR.svg = '<svg width="3%" height="3%" viewBox="0 0 640 480">\n  <g fill-rule="evenodd" stroke-width="1pt">\n    <path fill="#fff" d="M0 0h640v480H0z"/>\n    <path fill="#00267f" d="M0 0h213.3v480H0z"/>\n    <path fill="#f31830" d="M426.7 0H640v480H426.7z"/>\n  </g>\n</svg>\n';

    let cH = chord.createChordNode('CH');
    cH.description = 'Switzerland';
    cH.color = Color.red;
    cH.svg = '<svg width="3%" height="3%" viewBox="0 0 640 480">\n  <g fill-rule="evenodd" stroke-width="1pt">\n    <path fill="#d52b1e" d="M0 0h640v480H0z"/>\n    <g fill="#fff">\n      <path d="M170 195h300v90H170z"/>\n      <path d="M275 90h90v300h-90z"/>\n    </g>\n  </g>\n</svg>\n';

    let nL = chord.createChordNode('NL');
    nL.description = 'Netherlands';
    nL.color = Color.yellow;
    nL.svg = '<svg width="3%" height="3%" viewBox="0 0 640 480">\n  <path fill="#21468b" d="M0 0h640v480H0z"/>\n  <path fill="#fff" d="M0 0h640v320H0z"/>\n  <path fill="#ae1c28" d="M0 0h640v160H0z"/>\n</svg>\n';

    //#end region

    //#end region ---------------------------------------------------------

    return root;
};
```
