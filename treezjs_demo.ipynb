```python
%%javascript

/*
This file demonstrates the usage of treezjs with JupyterNotebooks.
Also see https://github.com/stefaneidelloth/treezjs
    
This might only be interesting for advanced users. 
Please first work through the tutorials under gettting_started_jupyter.

In order to use this treezjs demo, you need to open it as JupyterNotebook 
instead of using JupyterLab!

Switching to JupyterNotebooks can be done by changing the url
http://tatooine:8888/user/test/lab
to
http://tatooine:8888/user/test/tree

Or use direct url:
http://tatooine:8888/user/test/notebooks/Untitled.ipynb
*/
import Root from '../files/files/treezjs/src/root/root.js';
import Color from '../files/files/treezjs/src/components/color/color.js';

window.createModel = function(){

    var root = new Root('root');

    //#region MODELS ------------------------------------------------------
    var models = root.createModels('models');

    //#region executable

    var executable = models.createExecutable('executable');
    executable.executablePath = 'notepad.exe';
    executable.inputArguments = '';
    executable.inputPath = '';
    executable.outputArguments = '';
    executable.outputPath = '';

    //#end region

    //#end region ---------------------------------------------------------

    //#region STUDIES -----------------------------------------------------
    var studies = root.createStudies('studies');
    //#end region ---------------------------------------------------------

    //#region RESULTS -----------------------------------------------------
    var results = root.createResults('results');

    //#region page

    var page = results.createPage('page');
    page.width = '15 cm';
    page.height = '15 cm';
    page.color = Color.white;
    page.isHidden = false;

    var graph = page.createGraph('graph');
    graph.createXy('xy');

    //#end region

    //#end region ---------------------------------------------------------

    return root;
};
```
